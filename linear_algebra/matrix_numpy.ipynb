### 0~9 이미지 데이터 출력

```python
from sklearn.datasets import load_digits

digits = load_digits() # 데이터 로드
samples = [0,10,20,30,1,11,21,31] #선택된 이미지번호
d =[]
for i in range(8):
    d.append(digits.images[samples[i]])
d    
plt.figure(figsize=(8,2))
for i in range(8):
    plt.subplot(1, 8, i+1)
    plt.imshow(d[i], interpolation='nearest', cmap=plt.cm.bone_r)
    plt.grid(False); plt.xticks([]); plt.yticks([])
    plt.title('image {}'.format(i+1))

plt.suptitle('숫자 0과 1이미지')
plt.tight_layout()
plt.show()
```

```python
v = []
for i in range(8):
    v.append(d[i].reshape(64,1)) # 벡터화 64-1크기 벡터
    
plt.figure(figsize=(8,3))
for i in range(8):
    plt.subplot(1,8, i+1)
    plt.imshow(v[i], aspect=0.4,
              interpolation='nearest', cmap = plt.cm.bone_r)
    plt.grid(False); plt.xticks([]); plt.yticks([])
    plt.title('벡터 {}'.format(i+1))
    
plt.suptitle('벡터화된 이미지', y=1.05)
plt.tight_layout(w_pad=7)
plt.show()
```

### 벡터와 벡터의 곱셈, 벡터의 내적(가중합, 가중평균, 유사도)
- 벡터의 내적, dot product라고도 함.
- 두 벡터의 차원(길이)이 같아야한다.
- 앞으 벡터가 행벡터, 뒤의 벡터가 열 벡터여야 한다

```python
x = np.array([[1],[2],[3]])
y = np.array([[4],[5],[6]])
```

```python
(x.T @ y)[0][0] # 값을 수동으로 뽑아내주어야함
```

### 가중합(weighted sum)
- 예를들어 물건을 사는데 물건의 수량과 가격이 있음. 그것은 수량x가격이 있음
- 가중치가 있는 가중합
- 가중치를 모두 1로 만들면 일반적인 합(sum)을 계산하는 것임

```python
# 연습문제 2.2.1
p = np.array([[100],[80],[50]])
n = np.array([[3],[4],[5]])
(p.T@n)[0][0]
```

### 선형회귀 모형은 예측을 뽑아내는 모델
- x는 입력하는 숫자, y는 뽑아내려는 숫자
- 모델의 출력은 캐럿이라고 하는 기호를 붙여서 와이 햇이라고 읽음

```python
### 행렬의 곱셈
- 앞 행렬의 열과 뒤 행렬의 행과 같아야 한다.
```

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[1, 2], [3, 4], [5, 6]])
B@A
```

```python
X = np.array([[1],[2],[3]])
print((X.T)@X)
print(X@(X.T))
```
